# Q2) What are the percentage of restaurants by Dining Rating in Chennai?

```python
import numpy as np
diningRate=raw_df['Dining Rating'].value_counts().reset_index()
diningRate.rename(columns = {'Dining Rating': 'Dining Rating' , 'count':'Number of restaurants'},inplace=True)
```

```python
conditions = [
    (diningRate['Dining Rating'] >= 0) & (diningRate['Dining Rating'] < 2),
    (diningRate['Dining Rating'] >= 2) &  (diningRate['Dining Rating'] < 3),
    (diningRate['Dining Rating'] >= 3) & (diningRate['Dining Rating'] < 4),
    (diningRate['Dining Rating'] >= 4)
]

values = ['Very Poor','Poor', 'Average', 'Good']

diningRate['Dining Rating Type']= np.select(conditions, values, default='Unknown')


diningRate.sort_values(by='Dining Rating',ascending=False)
```

```python
fig=px.pie(diningRate,names='Dining Rating Type',values='Number of restaurants',
        title='Percentage of restaurants by Dining Rating',color='Dining Rating Type').update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

# Q10) Find the list of Affordable Restaurants?

The criteria for Affordable Restaurants would be:-
1) Low Price
2) High Rate.

First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.

Let me explain:-The most expensive restaurant has an average meal cost= 6000.
We'll try to stay economical and only pick the restaurants that are 1/4th of 6000.

```python
max_price=raw_df['Price for 2'].max()
max_price
```

```python
low_price=max_price/4
low_price
```

```python
low_price_df=raw_df[raw_df['Price for 2']<=low_price]
low_price_df
```

high_rate_df=raw_df[raw_df['Dining Rating']>=4]
high_rate_df

```python
lowcost_highrating_df=pd.merge(low_price_df,high_rate_df, how='inner',on=['Name of Restaurant','Location'])
```

```python
lowcost_highrating_df=lowcost_highrating_df[['Name of Restaurant','Location','Cuisine_x',
                                             'Price for 2_x','Dining Rating_x','Dining Rating Count_x']]
lowcost_highrating_df.rename(columns= {'Cuisine_x' : 'Cuisine','Price for 2_x' : 'Price for 2' ,
                                       'Dining Rating_x' : 'Dining Rating' , 
                                       'Dining Rating Count_x' :'Dining Rating Count'})

lowcost_highrating_df.head(10)                                           
```
